```python
import gym
env = gym.make("Taxi-v3") # making an environment  of Taxi-v3 where -v3 means version3 
rewards=[] #creating a list of rewards ,  show how many have been scored at the end of the games 
observation = env.reset()   #after the end of the game it get reseted 
for _ in range(50):  
    env.render()  # render  is a function in gym where it means display on to the screen. These function is used to diplay "Taxi -v3 model".
    action = env.action_space.sample()   # random sample of space is give it .
    obersvation , reward,done,info = env.step(action)
    rewards.append(reward)
env.close()    
```
